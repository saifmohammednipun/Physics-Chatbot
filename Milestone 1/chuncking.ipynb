# A comparative analysis of different chunking strategies

## Textbook Load

```python
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader('/Users/saifmohammed/Desktop/Physics-Chatbot/Text Book/Physics Classes 9-10 (English Version) - National Curriculum and Textbook Board of Bangladesh - PDF Room.pdf')
document = loader.load()
document
```

## Different Chuncking Strategies

### Fixed-size chunking

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size=500,chunk_overlap=100)
CTS_Chunked = text_splitter.split_documents(document)

for chunk in CTS_Chunked[20:25]: 
    print(chunk.page_content + "\n")
```

### Tokenized Chuncking

```python
from langchain.text_splitter import TokenTextSplitter
text_splitter = TokenTextSplitter(chunk_size=500, chunk_overlap=100)
token_chunked = text_splitter.split_documents(document)

for chunk in token_chunked[20:25]:
    print(chunk.page_content + "\n")
```

## Recursive Chuncking

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
recursive_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100, separators=['\n', '.', '!', '?'])
recursive_chunked = recursive_splitter.split_documents(document)

for chunk in recursive_chunked[20:25]:
    print(chunk.page_content + "\n")
```

## Specialized Chuncking

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# Initialize Recursive Character Text Splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500, 
    chunk_overlap=100  
)

# Split the document into chunks
chunks = text_splitter.split_documents(document)

# Verify the chunked output (e.g., show the first 5 chunks)
for chunk in chunks[5:10]:
    print(chunk.page_content + "\n")
```
